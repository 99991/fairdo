```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# encode non-numeric columns to labels
le = LabelEncoder()
iris_transform = iris.apply(le.fit_transform)

# make species binary
iris_transform['species'] = iris_transform['species'] == 0

# make label
iris_transform['petal_length'] = iris_transform['petal_length'] > 9
```

```python
iris_transform.head()
```

```python
n_neighbors = 5
```

```python
nbrs = NearestNeighbors(n_neighbors=n_neighbors, algorithm='ball_tree')
nbrs.fit(iris_transform)
indices = nbrs.kneighbors(iris_transform, return_distance=False)
```
